```python
def keywordSearch(data,data_dict,keywords):
    res = {}
    for i in range(data.shape[1]):
        
        temp = data_dict.loc[data.iloc[:,i].name,'name']
        temp = temp.strip().lower()
        score = 0
        for word in keywords.keys():
            if word in temp:
                score += keywords[word]  
        res[data.iloc[:,i].name]=score
    return res         
```

```python
def corr_hist_plt(data):
    # Create correlation matrix
    corr_matrix = data.corr().abs()

    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    
    plt.hist(upper.stack().values) # unique pairs of features
    plt.show()

def corr_feature_kept(data,cutoff):
    # Create correlation matrix
    corr_matrix = data.corr().abs()

    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

    # Find index of feature columns with correlation greater than 0.95
    to_drop = [column for column in upper.columns if any(upper[column] > cutoff)]
    print('number of features kept:',len(data.columns)-len(to_drop))
    return to_drop

def corr_feature_remove(data,to_drop): 
    df = data
    df = df.drop(columns=to_drop,axis=1)
    df.plot()
    return df
```
